```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

# Climate Change time data

```python
time_data=pd.read_csv('./climate_change.csv')
```

```python
time_data
for i in range(time_data.shape[0]):
    time_data.loc[i, 'date']=datetime.datetime.strptime(time_data.loc[i, 'date'], '%Y-%m-%d')

time_data.set_index(time_data['date'], inplace=True)
```

```python
time_data
```

```python
plt.figure(figsize=(25, 25))

fig, ax = plt.subplots()

ax.plot(time_data.index, time_data['co2'], color='b')

ax.set_xlabel('Time')
ax.set_ylabel('CO2', color='b')
ax.tick_params('y', colors='b')
ax2=ax.twinx()

ax2.plot(time_data['date'], time_data['relative_temp'], color='r')
ax2.set_ylabel('Temp', color='r')
ax2.tick_params('y', colors='r')

ax.set_title('change of CO2 and Temperature')
plt.show()
```

```python
time_data
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(time_data.index, time_data['co2'], color='b')
plt.xlabel('Time')
plt.ylabel('Co2')
plt.title('change of co2')
plt.xticks(rotation=90)

plt.subplot(2, 1, 2)
plt.plot(time_data.index, time_data['relative_temp'], color='r')
plt.xlabel('Time')
plt.ylabel('relative_temp')
plt.title('change of relative temp')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
subdata=time_data['1958-03-06':'1968-03-06']
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(subdata.index, subdata['co2'], color='b')
plt.xlabel('Time')
plt.ylabel('Co2')
plt.title('change of co2 in 1958~1968')
plt.xticks(rotation=90)

plt.subplot(2, 1, 2)
plt.plot(subdata.index, subdata['relative_temp'], color='r')
plt.xlabel('Time')
plt.ylabel('relative_temp')
plt.title('change of relative temp in 1958~1968')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

# MPG Data

```python
time_data2=pd.read_csv('./mpg.csv')
time_data2
```

```python
sns.relplot(x='model_year', y='horsepower', data=time_data2, kind='scatter')
plt.xlabel('year')
plt.ylabel('horsepower')
plt.show()

```

```python
sns.set_palette('Set1')
sns.relplot(x='model_year', y='horsepower', data=time_data2, kind='line', hue='origin', ci=None)
plt.xlabel('year')
plt.ylabel('horsepower')
plt.title('model_year vs horsepower')
plt.show()
```

```python
g=sns.relplot(x='model_year', y='horsepower', data=time_data2, kind='line', hue='origin', col='origin', ci='sd')
g.fig.suptitle('model_year vs horsepower based on origin', x=0.5, y=1.1)
plt.show()
```

```python
sns.catplot(x='origin', y='horsepower', data=time_data2, kind='bar', ci='sd')

plt.title('mean and std of horsepower based on origin')

plt.show()
```

```python
sns.catplot(x='origin', y='horsepower', data=time_data2, kind='box')
plt.title('box plot of horsepower based on origin')
plt.show()
```

```python
sns.catplot(x='origin', y='horsepower', data=time_data2, kind='point', color='blue')
plt.title('point plot of horsepower based on origin')
plt.show()
```

```python
sns.relplot(x='horsepower', y='mpg', data=time_data2, hue='origin', style='origin')
plt.title('horsepower vs mpg')
plt.show()
```

```python
sns.catplot(x='cylinders', kind='count', data=time_data2)
plt.title('number of cylinders')
```

```python
sns.catplot(x='cylinders', kind='count', data=time_data2, hue='origin')
plt.title('number of cylinders')
```

```python
count=time_data2[['origin', 'cylinders']]
count1=time_data2['origin']
count2=time_data2['cylinders']
```

```python
usa=count[count1=='usa']
japan=count[count1=='japan']
europe=count[count1=='europe']

usa_size=usa.groupby('cylinders').count()['origin']
japan_size=japan.groupby('cylinders').count()['origin']
europe_size=europe.groupby('cylinders').count()['origin']
```

```python
usa_size=np.array([0, 72, 0, 74, 0, 103])
japan_size=np.array([4, 69, 0, 6, 0, 0])
europe_size=np.array([0, 63, 3, 4, 0, 0])
```

```python
fig, ax=plt.subplots()

ax.bar([3, 4, 5, 6, 7, 8], usa_size, label='usa')
ax.bar([3, 4, 5, 6, 7, 8], japan_size, bottom=usa_size, label='japan')
ax.bar([3, 4, 5, 6, 7, 8], europe_size, bottom=(usa_size+japan_size), label='europe')

ax.set_xlabel('number of cylinders')
ax.set_ylabel('number of cars')
ax.set_title('barplot of cylinders based on origin')
plt.legend()
plt.show()
```

```python
fig, ax=plt.subplots()
ax.hist(time_data2['mpg'], bins=15)
ax.set_xlabel('mpg')
ax.set_title('histogram of MPG')

plt.show()
```

```python
sns.distplot(time_data2['mpg'], bins=20)
plt.title('histogram and density plot of MPG')
plt.show()
```

```python
corr=time_data2.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, linewidths=0.5, cbar='False', cmap='Blues')
plt.title('correlation matrix of mpg data', x=0.6, y=1.5)
```

```python
g=sns.pairplot(data=time_data2, diag_kind='kde', hue='origin', vars=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'])
g.fig.suptitle('pairplot of mpg data', y=1.05, size=20)
plt.show()
```
